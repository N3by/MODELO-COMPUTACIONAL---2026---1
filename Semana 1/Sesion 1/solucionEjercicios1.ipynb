```python
# Solución ejercicio 3 — término k²
def termino_cuadrado(n):
    resultado = n ** 2
    return resultado

print(termino_cuadrado(5))

f = lambda x: x ** 2
print(f(5))
```

```python
# Respuesta punto 4 (término al cubo)
def termino_cubo(num):
    k = num ** 3
    return k

print(termino_cubo(5))

tres = lambda x: x ** 3
print(tres(5))
```

```python
# RTA PUNTO 5 — potencias de 2

def potencia_dos(n):
    valor = 2 ** n
    return valor

print(potencia_dos(5))

tres = lambda x: 2 ** x
print(tres(5))
```

```python
# Ej 6 — inverso 1/k
def inverso(num):
    k = 1 / num
    return k

print(inverso(5))

tres = lambda x: 1 / x
print(tres(5))
```

```python
# Solución 7 — expresión (3k + 2)
def termino_lineal(num):
    k = (3 * num) + 2
    return k

print(termino_lineal(5))

tres = lambda x: (3 * x) + 2
print(tres(5))
```

```python
# Punto 8 — término k² − k
def numtres(num):
    k = (num ** 2) - num
    return k

print(numtres(5))

tres = lambda x: (x ** 2) - x
print(tres(5))
```

```python
# Respuesta ej. 9 — fracción k/(k+1)
def numtres(num):
    k = num / (num + 1)
    return k

print(numtres(5))

tres = lambda x: x / (x + 1)
print(tres(5))
```

```python
# Ej 10 — término (−1)^k
def signo_alternado(num):
    k = (-1) ** num
    return k

print(signo_alternado(2))

tres = lambda x: (-1) ** x
print(tres(2))
```

```python
# Solución 11 — (2k)²
def termino_doble_cuad(num):
    k = (2 * num) ** 2
    return k

print(termino_doble_cuad(2))

tres = lambda x: (2 * x) ** 2
print(tres(2))
```

```python
# Punto 12 — 1/2^k
def medio_potencia(num):
    k = 1 / (2 ** num)
    return k

print(medio_potencia(2))

tres = lambda x: 1 / (2 ** x)
print(tres(2))
```

```python
# Respuesta 13 — k(k+1)
def numtres(num):
    k = num * (num + 1)
    return k

print(numtres(2))

tres = lambda x: x * (x + 1)
print(tres(2))
```

```python
import math
# Ej 14 — factorial k!
def factorial_manual(num):
    k = 1
    for i in range(1, num + 1):
        k *= i
        print(k)

print("prueba", factorial_manual(3))

tres = lambda x: math.factorial(x)
print(tres(3))
```

```python
# Punto 15 — ln(k)
import math
def log_term(num):
    k = math.log(num)
    return k

print(log_term(2))

tres = lambda x: math.log(x)
print(tres(2))
```

## II. Ejercicios de Diccionarios en Python

1. Cree un diccionario con los nombres de cinco estudiantes como claves y sus calificaciones finales como valores. Calcule el promedio.
2. Dado un diccionario de productos y precios, incremente todos los precios en un 10%.
3. Escriba un programa que determine la clave asociada al valor máximo en un diccionario numérico.
4. Invierta las claves y valores de un diccionario (asuma que los valores son únicos).
5. Dado un diccionario con palabras y su frecuencia, elimine las palabras con frecuencia menor que 3.
6. Combine dos diccionarios sumando los valores de las claves comunes.
7. Ordene un diccionario por sus valores de forma ascendente.
8. Verifique si una clave ingresada por el usuario existe en un diccionario dado.

```python
# Diccionario estudiantes — calificaciones y promedio

calificaciones = {
    "Juan": [2, 2, 3],
    "Pedro": [4, 4, 3],
    "David": [5, 2, 3]
}

for nombre, notas in calificaciones.items():
    promedio = sum(notas) / len(notas)
    print(nombre, "promedio:", promedio)
```

```python
# Productos y precios — incremento 10%
precios = {
    "Comida": 20,
    "Ropa": 15,
    "Lavado": 30
}

for articulo, valor in precios.items():
    porcentaje = (10 * valor) / 100
    valor += porcentaje
    print(articulo, "precio final:", valor)
```

```python
# Clave con valor máximo
dic = {
    "C": 20,
    "R": 15,
    "L": 30,
    "M": 25
}

clave_max = max(dic, key=dic.get)
print("Llave del máximo:", clave_max)
print("Valor máximo:", dic[clave_max])
```

```python
# Invertir claves y valores (valores únicos)
dic = {
    "C": 20,
    "R": 15,
    "L": 30,
    "M": 25
}
# Nuevo diccionario: valor pasa a ser llave y llave a valor
dic_invertido = {valor: llave for llave, valor in dic.items()}

print(dic_invertido)
```

```python
# Palabras y frecuencia — quitar las que tienen freq < 3
frecuencias = {
    "hola": 5,
    "adiós": 2,
    "perro": 3,
    "gato": 1,
    "Pajaro": 6
}

claves_a_quitar = []
for palabra, freq in frecuencias.items():
    if freq < 3:
        claves_a_quitar.append(palabra)

for palabra in claves_a_quitar:
    del frecuencias[palabra]

print(frecuencias)    
```

```python
# Combinar dos diccionarios — sumar valores en claves comunes
dic = {"C": 20, "R": 15, "L": 30, "M": 25}
dic2 = {"C": 40, "X": 10, "Ñ": 30, "M": 5}

resultado = {}
for clave, valor in dic.items():
    resultado[clave] = valor
for clave, valor in dic2.items():
    if clave in resultado:
        resultado[clave] += valor
    else:
        resultado[clave] = valor

print(resultado)
```

```python
# Ordenar diccionario por valores (ascendente)
dic = {
    "C": 20, "R": 15, "L": 30, "M": 25,
    "T": 40, "X": 10, "Ñ": 30, "U": 5
}
# sorted por el valor (item[1]); dict() arma de nuevo el diccionario
ordenado = dict(sorted(dic.items(), key=lambda item: item[1]))
ordenado
```

```python
# Verificar si una clave existe; si no, pedir valor y agregar (salir con 'x')
dic = {
    "C": 20, "R": 15, "L": 30, "M": 25,
    "T": 40, "X": 10, "Ñ": 30, "U": 5
}

while True:
    llave = input("Ingresar llave: ")
    if llave in dic:
        print("Esa llave ya está en el diccionario")
        break
    if llave.lower() == "x":
        break
    valor = input("Ingresar valor: ")
    dic[llave] = valor

print("Diccionario final:", dic)
```

## III. Ejercicios de Multiplicación de Diccionarios en Python

(Se entiende por multiplicación la operación valor a valor entre diccionarios con claves comunes).

1. Dados dos diccionarios con las mismas claves y valores numéricos, construya un tercer diccionario donde cada valor sea el producto de los valores correspondientes.
2. Multiplique todos los valores de un diccionario por un escalar dado.
3. Dado un diccionario que representa dos vectores con las mismas claves, calcule su producto punto:
   
   $$
   \sum_{i} v_i \cdot w_i
   $$

4. Dado un diccionario de precios y otro de cantidades, calcule el costo total por producto y el costo total general.


```python
# Producto valor a valor (mismas claves)
dic = {"C": 20, "R": 15, "L": 30, "M": 25}
dic2 = {"C": 20, "R": 15, "L": 30, "M": 25}

producto = {}
for clave in dic:
    producto[clave] = dic[clave] * dic2[clave]

print(producto)
```

```python
# Multiplicar todos los valores por un escalar
dic = {"C": 20, "R": 15, "L": 30, "M": 25}
escalar = 2
resultado = {clave: dic[clave] * escalar for clave in dic}
print(resultado)
```

```python
# Producto punto de dos vectores (mismas claves)
v = {"C": 2, "R": 3, "L": 4, "M": 5}
w = {"C": 2, "R": 3, "L": 4, "M": 5}

dot = 0
for clave in v:
    dot += v[clave] * w[clave]

print("Producto punto:", dot)
```

```python
# Precios y cantidades — costo por producto y total
cantidades = {"Carro": 2, "Rojo": 3, "Llave": 4, "Malla": 5}
precios = {"Carro": 20, "Rojo": 30, "Llave": 40, "Malla": 50}

total_general = 0
for producto, cant in cantidades.items():
    costo_item = cant * precios[producto]
    print(f"Costo de {producto}: {costo_item}")
    total_general += costo_item

print("Costo total general:", total_general)
```
